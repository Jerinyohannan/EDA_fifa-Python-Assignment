```python
import matplotlib.pyplot as plt
import pandas as pd
```

1.Which country has the most number of players (score :1)


```python
df = pd.read_csv('fifa_data.csv')
country_player_count = df['Nationality'].value_counts()
most_players_country = country_player_count.index[0]
print(f"The country with the most players is: {most_players_country}")
```

2.Plot a bar chart of 5 top countries with the most number of players. (score :1)


```python
top_5_countries = country_player_count.head(5)
top_5_countries.plot(kind='bar', color='red')
plt.xlabel('Country')
plt.ylabel('Number of Players')
plt.title('Top 5 Countries with the Most Number of Players')
plt.show()
```

3.Which player has the highest salary? (score :1)


```python
df['Wage'] = df['Wage'].str.replace('€', '').str.replace('K', '000').astype(float)
```

```python
df_sorted_by_wage = df.sort_values(by='Wage', ascending=False)
highest_salary_player = df_sorted_by_wage.loc[df_sorted_by_wage['Wage'].index[0]]
print(f"The player with highest salary is: {highest_salary_player['Name']} with a salary of {highest_salary_player['Wage']}")
```

4.Plot a histogram to get the salary range of the players. (score :1)


```python
plt.hist(df['Wage'], bins=20,color='red',edgecolor='black')
plt.xlabel('Salary (in Euros)')
plt.ylabel('Number of Players')
plt.title('Salary Range of FIFA Players')
plt.show()
```

5.Who is the tallest player in the fifa? (score :1)


```python
df['Height'] = df['Height'].str.replace("'", '.').astype(float)
```

```python
df_sorted_by_height = df.sort_values(by='Height', ascending=False)
tallest_player = df_sorted_by_height.loc[df_sorted_by_height['Height'].index[0]]
print(f"The Tallest player in FIFA is {tallest_player['Name']} with a height of {tallest_player['Height']}")
```

6.Which club has the most number of players? (score :1)


```python
club_player_count = df['Club'].value_counts()
most_players_club = club_player_count.index[0]
print(f"The club with the most players is: {most_players_club}")
```

7.Which foot is most preferred by the players?Draw a bar chart for preferred foot (score :1)

```python
preferred_foot_count = df['Preferred Foot'].value_counts()
preferred_foot_count.plot(kind='bar', color='red')
plt.xlabel('Preferred Foot')
plt.ylabel('Number of Players')
plt.title('Preferred Foot of FIFA Players')
plt.show()
```
